```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('flight_data.csv')

# Convert 'FlightDateTime' to datetime format
df['FlightDateTime'] = pd.to_datetime(df['FlightDateTime'])

# Extract hour from 'FlightDateTime'
df['DepartureHour'] = df['FlightDateTime'].dt.hour

# Group by 'DepartureHour' and calculate average quantity
avg_quantity_per_hour = df.groupby('DepartureHour')['Quantity'].mean()

# Plot the graph for Average Quantity per Departure Hour
plt.figure(figsize=(10, 6))
avg_quantity_per_hour.plot(kind='bar', color='skyblue')
plt.title('Average Quantity per Departure Hour')
plt.xlabel('Departure Hour')
plt.ylabel('Average Quantity')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Convert 'FlightDurationInHours' to half-hour segments
df['FlightHalfHourSegment'] = (df['FlightDurationInHours'] * 2).astype(int) / 2

# Group by 'FlightHalfHourSegment' and calculate average quantity
avg_quantity_per_flight_time = df.groupby('FlightHalfHourSegment')['Quantity'].mean()

# Plot the graph for Average Quantity per Flight Time
plt.figure(figsize=(10, 6))
avg_quantity_per_flight_time.plot(kind='bar', color='green')
plt.title('Average Quantity per Flight Time (Half-Hour Segments)')
plt.xlabel('Flight Time (Half-Hour Segments)')
plt.ylabel('Average Quantity')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Print the size of each group from flight time
#flight_time_group_sizes = df.groupby('FlightHalfHourSegment').size()
#print("Size of each group from flight time:")
#print(flight_time_group_sizes)

# Group by 'Month' and calculate average quantity
avg_quantity_per_month = df.groupby('Month')['Quantity'].mean()

# Plot the graph for Average Quantity per Month
plt.figure(figsize=(10, 6))
avg_quantity_per_month.plot(kind='bar', color='purple')
plt.title('Average Quantity per Month')
plt.xlabel('Month')
plt.ylabel('Average Quantity')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Group by 'WeekNum' and calculate sum or mean of quantity
avg_quantity_per_week = df.groupby('WeekNum')['Quantity'].mean()

# Plot the graph for Quantity per Week Number
plt.figure(figsize=(10, 6))
avg_quantity_per_week.plot(kind='bar', color='purple')
plt.title('Average Quantity per Week Number')
plt.xlabel('Week Number')
plt.ylabel('Average Quantity')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()


# Group by 'day_of_week' and calculate average quantity
avg_quantity_per_weekday = df.groupby('day_of_week')['Quantity'].mean()

# Plot the graph for Average Quantity per Weekday
plt.figure(figsize=(10, 6))
avg_quantity_per_weekday.plot(kind='bar', color='orange')
plt.title('Average Quantity per Weekday')
plt.xlabel('Weekday')
plt.ylabel('Average Quantity')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Group by 'HolidayInDestination' and calculate average quantity
avg_quantity_per_holiday_destination = df.groupby('HolidayInDestination')['Quantity'].mean()

# Plot the graph for Average Quantity per Holiday in Destination
plt.figure(figsize=(8, 5))
avg_quantity_per_holiday_destination.plot(kind='bar', color='blue')
plt.title('Average Quantity per Holiday in Destination')
plt.xlabel('Holiday in Destination')
plt.ylabel('Average Quantity')
```
